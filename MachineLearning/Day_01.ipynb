```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.rand(50, 1)
Y = 1.8*(X**5) -\
    1.6*(X**4) +\
    0.9*(X**3) +\
    0.5*(X**2) +\
    0.3*X + 0.1*np.random.rand(50, 1)
plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
fig = plt.figure()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
for i in range(1, 6):
    polynomial = PolynomialFeatures(i)
    x_poly = polynomial.fit_transform(X)
    scalar = StandardScaler()
    scalar.fit(x_poly)
    x_stnd = scalar.transform(x_poly)
    model = LinearRegression()
    model.fit(x_stnd, Y)
    output = model.predict(x_stnd)
    plt.scatter(X, output, label= "predict")
    plt.scatter(X,Y,label="given")
    plt.show()
```
