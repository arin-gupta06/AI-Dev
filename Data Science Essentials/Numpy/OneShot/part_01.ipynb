### Array Reshaping

```python
arr1 = np.arange(12)
print("Original array:\t", arr1)
```

```python
reshape = arr1.reshape([3,4])
print(reshape)
```

```python
flatened = reshape.flatten() #Returs the copy of the array
print(flatened)
```

```python
raveled = reshape.ravel() #Actual array
print(raveled)
```

```python
#Transpose
transpose = reshape.T
print(transpose)
```
